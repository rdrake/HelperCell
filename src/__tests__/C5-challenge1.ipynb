## The Challenge Exercise 1

**Background:** Library Late Fee Calculator

A local library charges late fees for overdue books based on how many days a book is overdue. The fee structure is as follows:

| Days Overdue   | Fee Per Book |
|---------------|-------------|
| 1 - 5 days    | \$0.50 per day |
| 6 - 10 days   | \$1.00 per day |
| 11 - 15 days  | \$1.50 per day |
| 16+ days      | \$2.00 per day (plus a warning for suspension) |






Your program should: 
1. Ask the user how many books they returned late.
2. Ask how many days overdue the books are.
3. Calculate the total late fee based on the table above.
4. Display the total fee and, if the books are overdue by **16 or more days**, prints a warning message about possible account suspension.

**Sample test case 1 input**

**Sample test case 1 output**

```
    You have 3 overdue books for 7 days. The total late fee is: $21.00
```


**Sample test case 2 input**

**Sample test case 2 output**

```
  You have 2 overdue books for 18 days. The total late fee is: $72.00
  WARNING: Your account may be suspended due to excessive late returns!
```

**Tips/Strategies**
- You were given the code to take the user's input already.

````python
# reading input from user
# reading the number of books and late days as a int
nbooks = int(input("Enter the number of overdue books:"))
ndays = int(input("Enter the number of days overdue:"))

if ndays <= 5: #calculates the late fee for books 1 to 5 days late and prints amount owing
    print("You have", nbooks, "books overdue for", ndays, "days. The total late fee is: $", nbooks*0.50*ndays)  
elif ndays <=10: #calculates the late fee for books 6 to 10 days late and prints amount owing
    print("You have", nbooks, "books overdue for", ndays, "days. The total late fee is: $", nbooks*1.00*ndays)
elif ndays <=15: #calculates the late fee for books 11 to 15 days late and prints amount owing
    print("You have", nbooks, "books overdue for", ndays, "days. The total late fee is: $", nbooks*1.50*ndays)
else: #calculates the late fee for books more than 15 days late. prints fees owing and a warning message.
    print("You have", nbooks, "books overdue for", ndays, "days. The total late fee is: $", nbooks*2.00*ndays)
    print("WARNING: Your account may be suspended due to excessive late returns!")
````

**Background:** 2-player rock, paper, scissors

You know that, to determine the winner you will apply the rules:
   - Rock beats Scissors

**Tips/Strategies**
