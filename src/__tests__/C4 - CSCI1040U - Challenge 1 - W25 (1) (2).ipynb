## The Challenge Exercise 1

**Background:** Library Late Fee Calculator

A local library charges late fees for overdue books based on how many days a book is overdue. The fee structure is as follows:

| Days Overdue   | Fee Per Book |
|---------------|-------------|
| 1 - 5 days    | \$0.50 per day |
| 6 - 10 days   | \$1.00 per day |
| 11 - 15 days  | \$1.50 per day |
| 16+ days      | \$2.00 per day (plus a warning for suspension) |






Your program should: 
1. Ask the user how many books they returned late.
2. Ask how many days overdue the books are.
3. Calculate the total late fee based on the table above.
4. Display the total fee and, if the books are overdue by **16 or more days**, prints a warning message about possible account suspension.

**Sample test case 1 input**

**Sample test case 1 output**

```
    You have 3 overdue books for 7 days. The total late fee is: $21.00
```


**Sample test case 2 input**

**Sample test case 2 output**

```
  You have 2 overdue books for 18 days. The total late fee is: $72.00
  WARNING: Your account may be suspended due to excessive late returns!
```

**Tips/Strategies**
- You were given the code to take the user's input already.

````python
# reading input from user
# reading the number of books and late days as a int
number_books = int(input("Enter the number of overdue books:"))
days_overdue = int(input("Enter the number of days overdue:"))

# Fee per book based on days overdue
if 1 <= days_overdue <= 5:
    fee_per_book = 0.50
elif 6 <= days_overdue <= 10:
    fee_per_book = 1.00
elif 11 <= days_overdue <= 15:
    fee_per_book = 1.50
else:
    fee_per_book = 2.00  # 16+ days

# Calculate Fee Due
total_fee = number_books * fee_per_book * days_overdue

# Output of overdue books and corresponding fee
print(f"Unfortunately you have {number_books} overdue books for {days_overdue} days. Therefore your total overdue fee is: ${total_fee}")

# Suspension if Applicable
if days_overdue >= 16:
    print("WARNING!! Your account is close to facing suspension due to a maximized number of overdue books and days. consider being more responsible you fool!")

````
